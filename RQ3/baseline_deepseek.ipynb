```python
import pandas as pd
import requests
from tqdm import tqdm

# Load the dataset
testingset = pd.read_csv("Sythetic_old_question.csv")




```

```python
testingset.head()
```

```python
# System prompt for DeepSeek baseline
system_prompt = """You are an expert programmer answering Stack Overflow questions. Provide clear, accurate, and concise answers to the Question and end with 'END_OF_ANSWER'."""

user_prompts = testingset["Paraphrased Question"].to_list()
response_list = []

DRY_RUN = False  # Set to False to enable real API calls
API_URL = "http://localhost:11434/api/generate"
MODEL_NAME = "deepseek-r1:70b"

def generate_deepseek(final_prompt):
    if DRY_RUN:
        return f"[DRY_RUN] Answer for: {final_prompt}"
    data = {
        "model": MODEL_NAME,
        "prompt": final_prompt,
        "stream": False
    }
    resp = requests.post(API_URL, json=data)
    return resp.json().get("response", "")

# Generate answers for each prompt
for user_prompt in tqdm(user_prompts, desc='DeepSeek Generating'):
    final_prompt = f"""{system_prompt}\n\nQuestion: {user_prompt}\n"""
    answer = generate_deepseek(final_prompt)
    response_list.append(answer)
```

```python
testingset['Generated Response'] = response_list
```
